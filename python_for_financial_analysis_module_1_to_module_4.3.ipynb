```python
#Using .shift() function to create a new column that shows the closing price of next day

fb['price1'] = fb['Close'].shift(-1)
fb.head()
```

```python
#Creating a new column priceDiff to calculate price diff in closing prices

fb['priceDiff'] = fb['price1'] - fb['Close']
fb.head()
```

```python
# Column to calculate the Daily Returns = Price Difference / Close

fb['dailyReturns'] = fb['priceDiff'] / fb['Close']
fb.head()
```

```python
# Column to calculate the direction of the stock price each day based on price difference

fb['direction'] = [1 if fb.loc[ ei, 'priceDiff'] > 0 else -1  for ei in fb.index]
fb.head()
```

```python
# Column to calculate the average of the Closing price over the last 3 days known as moving average

fb['Average3'] = (fb['Close'] + fb['Close'].shift(1) + fb['Close'].shift(2)) / 3
fb.head()
```

```python
# Column to calculate the moving average using rolling() function which can accept a int for number of days
# MA40 is average for last 40 days, while MA200 is average for last 200 days

fb['MA40'] = fb['Close'].rolling(40).mean()
fb['MA200'] = fb['Close'].rolling(200).mean()

fb.head()
```

```python
# PLOT the close and moveing average columns using plot function

fb['Close'].plot(legend = True)
fb['MA40'].plot(legend = True) # Represents Fast Signal
fb['MA200'].plot(legend = True) # Represents Slow Signal which represents long history

# if MA40 > MA200 some traders believe that the stock price will move upward for a while
```

```python
# Building a Trading Stratergy
# We will create 2 variables MA10 and MA50 
# if MA10 > MA50 buy and hold one share (long) else we will have 0 shares
# Create a new variable shares to determine weather to buy the share or not
# if share == 1, Profit = price1 - close else Profit = 0
# Plot profit 
# fb['wealth'] = fb['Profit'].cumsum() to calcuate proit/loss as the days go by
# check fb.tail() to determine the total profit made over the years 
# print investment and gain and plot wealth to show how wealth varied over the years


fb['MA10'] = fb['Close'].rolling(10).mean()
fb['MA50'] = fb['Close'].rolling(50).mean()
```

```python
fb['shares'] = [1 if fb.loc[ei, 'MA10'] > fb.loc[ei, 'MA50'] else 0 for ei in fb.index]
fb['shares1'] = [1 if fb.loc[ei, 'MA40'] > fb.loc[ei, 'MA200'] else 0 for ei in fb.index]
```

```python
fb['profit'] = [fb.loc[ei, 'price1'] - fb.loc[ei, 'Close'] if fb.loc[ei, 'shares'] == 1 else 0 for ei in fb.index]
fb['profit1'] = [fb.loc[ei, 'price1'] - fb.loc[ei, 'Close'] if fb.loc[ei, 'shares1'] == 1 else 0 for ei in fb.index]

fb = fb.fillna(0)
fb.head()
```

```python
fb['profit'].plot()
plt.axhline(y = 0, color = 'red')
```

```python
fb['profit1'].plot()
plt.axhline(y = 0, color = 'red')
```

```python
fb['wealth'] = fb['profit'].cumsum()
fb['wealth1'] = fb['profit1'].cumsum()

fb.tail()
```

```python
# using sampling with and without replacement

data = pd.DataFrame()
data['population'] = [12,45,23,76,89,96,56,54,90,91,94,14,15,11,16]
sample_without_replacement = data['population'].sample(5, replace = False)
sample_with_replacement = data['population'].sample(5, replace = True)
print( sample_without_replacement, sample_with_replacement )
```

```python
# Difference between parameter and statistics i.e. between population and sample

print( " population mean : ", data['population'].mean())
print( " population variance : ", data['population'].var(ddof = 0))
print( " population std : ", data['population'].std(ddof = 0))
print( " population shape : ", data['population'].shape[0])

a_sample = data['population'].sample(10, replace = True)

print( " sample mean : ", a_sample.mean())
print( " sample variance : ", a_sample.var(ddof = 1))
print( " sample std : ", a_sample.std(ddof = 1))
print( " sample shape : ", a_sample.shape[0])
```

```python
# Why the ddof changes in sample and population?

sample_length = 500
sample_variance_collection0= [data[ 'population' ].sample (50,replace=True).var(ddof=0) for i in range (sample_length) ]
sample_variance_collection1=[data['population'].sample (50,replace=True).var(ddof=1) for i in range (sample_length)]

print(data['population'].var(ddof = 0))
print(pd.DataFrame(sample_variance_collection0).mean())
print(pd.DataFrame(sample_variance_collection1).mean())

# Hence we can observe that the average using ddof = 1 is more close to the actual variance
```

```python
# Standardizaion, two and one tail test 

xbar = fb['logReturn'].mean()
s = fb['logReturn'].std(ddof = 1)
n = fb['logReturn'].shape[0]
zhat = (xbar - 0) / (s / (n ** 0.5))
print(zhat)
```

```python
# Calculating quantiles based on a given alpha Two tail testing

alpha = 0.05 # = 5%
zleft = norm.ppf(alpha/2, 0, 1)
zright = -zleft
print("Quantiles are : ", zleft, zright)
print("Should we Reject: ", zhat>zright or zhat<zleft)
```

```python
# One Tail Testing
print("Should we Reject: ", zhat>zright)
```

```python
# Calculating P-value method (most popular)
# Reject Null Hypothesis if p < alpha

p = 1 - (norm.cdf(abs(zhat), 0, 1))
print("Should we Reject: ", p < alpha)
```

```python
# Simple Linear Regression

b0 = 1
b1 = 2
housing['guess'] = b0 + b1 * housing['RM']

housing['error'] = housing['MEDV'] - housing['guess']
indices = [7,20,100]
print(housing['error'].loc[indices])
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols(formula = 'MEDV ~ RM', data = housing).fit()
b0 = model.params['Intercept']
b1 = model.params['RM']
housing['best_guess'] = b0 + b1 * housing['RM']
```

```python
plt.figure(figsize=(10, 10))

# Scatter plots
plt.scatter(housing['RM'], housing['MEDV'], color='g', label='Real')
#plt.scatter(housing['RM'], housing['y'], color='b', label='Model')

# Line plots
plt.plot(housing['RM'], housing['guess'], color='red', label='Guess')
plt.plot(housing['RM'], housing['best_guess'], color='yellow', label='Best Guess')

# Axis labels
plt.ylabel('MEDV / $1000')
plt.xlabel('RM / Number of Rooms')

# Axis limits
plt.xlim(np.min(housing['RM']) - 2, np.max(housing['RM']) + 2)
plt.ylim(np.min(housing['MEDV']) - 2, np.max(housing['MEDV']) + 2)

# Legend and show
plt.legend()
plt.show()
```

```python
housing['Error'] = housing['MEDV'] - housing['best_guess']

plt.figure(figsize = (16,9))
plt.title('Residual vs Order')
plt.plot(housing.index, housing['Error'], color = 'green')
plt.axhline( y = 0, color = 'red')
plt.grid(True)
plt.legend()
plt.show()
```
